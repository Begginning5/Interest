```python
# Method1: pdfplumber -> extract text from PDF files.
import string

import pdfplumber
# Extract PDF data
text = ""
path = "Business Observations - Sheet8.pdf"
with pdfplumber.open(path) as pdf:
   for page in pdf.pages:
      text += page.extract_text()

strings = text.split("\n")

i = 0
n = 0
while i < len(strings):
    strings[i] = strings[i].split()
    for n in range(len(strings[i])):
      if strings[i][n] == "cartons":
         strings[i][n] = 0
      if i > 0:
         strings[i][n] = float(strings[i][n])
      n += 1
    i += 1
df = pd.DataFrame(strings).fillna(0)

# for i in range(len(strings)):
   # print(len(strings[i]))
   # print(strings[i])
df

#----------------------------------------------
# So this method lost the datatype.
#----------------------------------------------
# method_1["line"] = method_1["line"].astype(str)

# method_1["line"] = (
   #  method_1["line"]
   #  .str.replace(",", "", regex=False)
   #  .str.replace("%", "", regex=False)
   #  .str.replace("  ", " ", regex=False)
   #  .str.strip()
# )

# method_1["line"] = pd.to_numeric(method_1["line"], errors="coerce")

# method_1["line"].dtype
# method_1["line"].head().to_list()
# text
```

```python
# Method4: pdf2image + pytesseract -> extract text from scanned PDF files.
from pdf2image import convert_from_path
import pytesseract

pages = convert_from_path("yourfile.pdf")

text = ""
for page in pages:
   text += pytesseract.image_to_string(page)

print(text)
```
